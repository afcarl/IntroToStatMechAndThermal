# Can we do the two-state paramagnet?

```python
Ntot = 200
Nup = np.arange(Ntot,-1,-1)
print Nup
```

```python
UmuB = Ntot - 2*Nup
MNmu = -UmuB/Ntot
Omega = sp.special.binom(Ntot,Nup)
Sk = np.log(Omega)
kTuB = (UmuB[2:]-UmuB[:-2])/(Sk[2:]-Sk[:-2])#dU/dS
CNk = (UmuB[4:]-UmuB[:-4])/(kTuB[2:]-kTuB[:-2])*Ntot #dU/NdT
```

```python
Sk
```

```python
plt.plot(kTuB[1:-1],CNk)
plt.axis([0,5,0,35000])
```

```python
plt.plot(UmuB,Sk)
plt.xlabel(r'$\frac{U}{\mu B}$')
plt.ylabel(r'$S/k$')
```

```python
plt.plot(UmuB[1:-1],kTuB)
plt.xlabel(r'$\frac{U}{\mu B}$')
plt.ylabel(r'$\frac{kT}{\mu B}$')
```
